```python
from numpy import *
from matplotlib.pyplot import *
```

```python
%matplotlib notebook
x = np.arange(-5,5,0.2)
y = np.arange(-4,4,0.2)

xx,yy =np.meshgrid(x,y)

def func(xx,yy):
    return exp(-(xx**2+yy**2))+exp(-((xx-2)**2+(yy-1)**2))+3*exp(-((xx+3)**2+(yy-1)**2))

subplots()
ax = gca()
ax.imshow(np.transpose(func(xx,yy)),aspect='auto',cmap='cividis_r',alpha=0.5)

data = func(xx,yy)

points=[]
colors=[]

#####################
def in_set(set, arr):
    for i in range(0,len(set)):
        if set[i][0] == arr[0] and set[i][1] == arr[1]:
            return True
    return False        

#####################

####initial grid points
step = 5
for j in range(0,len(data[1,:]),10):
    for i in range(0,len(data[:,j]),10):
        points.append(np.array([i,j]))
        colors.append('r')

threshold=0.5
for j in range(0,len(data[1,:]-1)):
    accum = 0
    for i in range(0,len(data[:,j])-1):
        accum += np.abs(data[i+1,j]-data[i,j])
        if accum > threshold:
            accum = 0
            if not in_set(points,np.array([i,j])):
                points.append(np.array([i,j]))
                colors.append('g')


for i in range(0,len(data[:,1]-1)):
    accum = 0
    for j in range(0,len(data[i,:])-1):
        accum += np.abs(data[i,j+1]-data[i,j])
        if accum > threshold:
            accum = 0
            if not in_set(points,np.array([i,j])):
                points.append(np.array([i,j]))
                colors.append('g')

            
points = np.array(points)
ax.scatter(points[:,0],points[:,1],edgecolor=colors[:],marker='o',s=10,facecolors='none',lw=1,alpha=0.7)
```
